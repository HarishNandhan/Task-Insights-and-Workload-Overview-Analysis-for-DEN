```python
# Trim whitespace from the "Assigned to" column and clean it as needed
ado['Customer Name'] = ado['Customer Name'].str.strip().str.lower()

# Get the unique categories in the "Assigned to" column and their value counts
cust_name_counts = ado['Customer Name'].value_counts()

# Display the number of unique categories and the counts for each category
num_categories = cust_name_counts.shape[0]
print(f"Number of unique categories in 'Cust Name': {num_categories}")
print("\nValue counts for each category:")
print(cust_name_counts)
```

```python
import pandas as pd

# Convert the "Created Date" column to datetime
ado['Created Date'] = pd.to_datetime(ado['Created Date'], errors='coerce')

# Extract only the date part and remove the time
ado['Created Date'] = ado['Created Date'].dt.date

# Display the cleaned DataFrame to verify the changes
ado.head()
```

```python
# Remove rows that are either before 2022-11-08 or after 2024-10-31 in "Created Date"
ado = ado[~((ado['Created Date'] < '2022-11-08') | (ado['Created Date'] > '2024-10-31'))]

# Display the first few rows to verify the changes
ado.head()
```

```python
# Trim whitespace from the "Assigned to" column and clean it as needed
ado['Assigned To'] = ado['Assigned To'].str.strip().str.lower()

# Get the unique categories in the "Assigned to" column and their value counts
developer_counts = ado['Assigned To'].value_counts()

# Display the number of unique categories and the counts for each category
num_categories = developer_counts.shape[0]
print(f"Number of unique categories in 'Developer': {num_categories}")
print("\nValue counts for each category:")
print(developer_counts)
```
