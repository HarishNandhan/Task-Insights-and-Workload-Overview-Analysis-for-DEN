```python
sn.drop(columns=['Assignment group','Application', 'Short description','Due date'], inplace=True)

# Display the cleaned dataframe to verify the changes
sn.head()
```

```python
# Convert the "Opened" and "Updated" columns to datetime if they are not already
sn['Opened'] = pd.to_datetime(sn['Opened'], errors='coerce')
sn['Updated'] = pd.to_datetime(sn['Updated'], errors='coerce')

# Extract only the date part and remove the time
sn['Opened'] = sn['Opened'].dt.date
sn['Updated'] = sn['Updated'].dt.date

# Display the cleaned dataframe to verify the changes
sn.head()
```

```python
# Convert "Azure DevOps Id" from float to integer
sn['Azure DevOps Id'] = sn['Azure DevOps Id'].fillna(0).astype(int)

# Display the cleaned dataframe to verify the changes
sn.head(100)
```

```python
sn.info()
```

```python
# Trim whitespace from the "Assigned to" column and clean it as needed
sn['Assigned to'] = sn['Assigned to'].str.strip().str.lower()

# Get the unique categories in the "Assigned to" column and their value counts
assigned_to_counts = sn['Assigned to'].value_counts()

# Display the number of unique categories and the counts for each category
num_categories = assigned_to_counts.shape[0]
print(f"Number of unique categories in 'Assigned to': {num_categories}")
print("\nValue counts for each category:")
print(assigned_to_counts)
```

```python
import numpy as np

# List of categories to use for random imputation
impute_categories = ['rachel', 'mel', 'harry']

# Identify rows with missing values in "Assigned to"
missing_indices = sn[sn['Assigned to'].isnull()].index

# Randomly assign values from impute_categories to the missing entries
np.random.seed(0)  # Set seed for reproducibility
sn.loc[missing_indices, 'Assigned to'] = np.random.choice(impute_categories, size=len(missing_indices))

# Display the updated "Assigned to" column to verify the imputation
print(sn['Assigned to'].value_counts())
```
